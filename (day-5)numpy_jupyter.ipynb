***1. write a program to find Fibonacci series***

```python
n = int(input("Enter the maximum value for the Fibonacci series: "))
a = 0
b = 1
print("Fibonacci series up to", n, ":")
while a <= n:
    print(a, end=" ")
    a, b = b, a + b
```

***3. write a program to calculate the gross income of a family***

```python
salary_1=int(input("Enter salary of person1: "))
salary_2=int(input("Enter salary of person2: "))
salary_3=int(input("Enter salary of person3: "))
tax=int(input("Enter tax amount: "))
expenses= int(input("Enter total expenses: "))
gross_income =salary_1+salary_2+salary_3
net_income =gross_income-tax-expenses
print("Gross Income:", gross_income)  #overall income that came
print("Net Income:", net_income)   #income left after necessary expenses
```

***4. write a program to make a calculator(keeps going until stopped)***

```python
continuation="y"
while continuation=="y":
    a=int(input("enter first number:"))
    b=int(input("enter second number:"))
    operation=input("enter the operation you want to perform(+,-,*,/):")
    if operation=="+":
        print("Result:", a + b)
    elif operation == "-":
        print("Result:", a - b)
    elif operation == "*":
        print("Result:", a * b)
    elif operation == "/":
        if b == 0:
            print("Invalid input: Cannot divide by zero.")
        else:
            print("Result:", a / b)
    else:
        print("invalid operation entered")
    continuation = input("enter y to continue, or any other key to exit: ")
```
